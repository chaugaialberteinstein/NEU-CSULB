## Chart 1: Plot displaying the inventory level of a specific item across all warehouses at a particular point in time

> We can see the total quantity of each warehouse using groupby function

```
df.groupby('Warehouse_block')[['Quantity']].sum()
```

```
df['Product_name'].value_counts().head(2)
```

> To see how many items of a specific product in each warehouse we just need to parse the name of items on 'product' variable

```


# Specify the product and the point in time you want to visualize
product = "S4S Stylish Women's Push-up Bra"

# Filter the DataFrame to get the inventory data for the specified product
filtered_df = df[df['Product_name'] == product]

# Group the filtered DataFrame by 'Warehouse' and calculate the sum of inventory for each warehouse
grouped_df = filtered_df.groupby('Warehouse_block')['Quantity'].sum()

# Plot the inventory levels
plt.bar(grouped_df.index, grouped_df.values)
plt.xlabel('Warehouse')
plt.ylabel('Inventory')
plt.title(f'Inventory Levels of {product} across Warehouses')
plt.show()
```

## Chart 2: Statistical chart of the quantity (or value) of goods imported/exported in all warehouses during a period.

```
df.groupby(['Warehouse_block']).agg({'Quantity':'sum'}).sort_values(by='Quantity',ascending=False).plot(kind='barh',legend=True,color='sandybrown')
# Set the plot title and labels
plt.title("Quantity by Warehouse Block")
plt.xlabel("Frequency")
plt.ylabel("Warehouse Block")

# Display the plot
plt.show()
```

```
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

grouped_df = filtered_df.groupby(['Warehouse_block', 'DELIVERED_TO_WAREHOUSE_DATE'])['Quantity'].sum()
grouped_df = grouped_df.reset_index().sort_values('DELIVERED_TO_WAREHOUSE_DATE')

plt.figure(figsize=(10, 6))  # Optional: Set the figure size

for warehouse, data in grouped_df.groupby('Warehouse_block'):
    data.plot(kind='line', x='DELIVERED_TO_WAREHOUSE_DATE', y='Quantity', label=warehouse)

plt.xlabel('DELIVERED_TO_WAREHOUSE_DATE')
plt.ylabel('Total Quantity')
plt.title('Total Quantity by Date Over Time in Each Warehouse')
plt.legend()

# Format the x-axis as dates
date_format = mdates.DateFormatter('%Y-%m-%d')  # Customize the date format as needed
plt.gca().xaxis.set_major_formatter(date_format)

plt.gcf().autofmt_xdate()  # Auto-format the x-axis date labels for better readability

plt.show()
```

## Chart 8 : "Difference in revenue from shipping orders / shipping costs in the year 2020"

```
fit_cost = df[['Revenue','Cost_of_the_Product_x','DELIVERED_DATE']]
fit_cost['dif'] = fit_cost['Revenue'] - fit_cost['Cost_of_the_Product_x']


plt.figure(figsize=(10,6))

fit_cost_2020 = fit_cost[fit_cost['DELIVERED_DATE'].dt.year == 2020]

# Create the stem plot
plt.stem(fit_cost_2020['DELIVERED_DATE'], fit_cost_2020['dif'])
plt.xlabel('Date')
plt.ylabel('Difference')
plt.title('Difference in revenue from shipping orders / shipping costs in the year 2020')
plt.show()
```

## Chart 4: Statistical chart of the number of orders by customers in a period of time

```
df4a = pd.read_csv('New VehicleJourneyorderinfor.csv')
df4a = df4a[['order_id_date']]
df4b = pd.read_csv('New VehicleJourney_Ecom_time.csv')
df4b = df4b[['customer_id']]
# Concat 2 data frame 
df4 = pd.concat([df4a, df4b], axis = 1)
df4
```

```
# Change column "order_id_date" to datetime
df4['order_id_date']= pd.to_datetime(df4['order_id_date'], dayfirst=True, errors='coerce')

# Sort the dataset in the ascending order of date
df4 = df4.sort_values(by = 'order_id_date')

# Set period
start_date = pd.to_datetime('2021-12-16')
end_date = pd.to_datetime('2022-02-16')
# Filtered the data frame
df4_filtered = df4[(df4['order_id_date'] >= start_date) & (df4['order_id_date'] <= end_date)]
print(df4_filtered)
```

```
#Plot
order_date_id = df4_filtered[['customer_id','order_id_date']].groupby(['order_id_date']).count()

plt.figure(figsize=(15, 6))

plt.plot(order_date_id.index, order_date_id['customer_id'], marker='o', markersize = 10, linewidth=3)


plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Customer over Time',fontsize=12, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

In general, the number of orders by customers over time fluctuated. However, as we can see from the graph, customer purchased more orders in December and January and the number of orders started to fall from February. This trend stands to common sense as the end of the year and the beginning of the next year is considered as holiday and shopping season.

##     Chart 6 : comparing the frequency of vehicle usage within a fleet during a period.

```
df6_1 = pd.read_csv('New Fleet Order.csv')
print(df6_1.columns.tolist())
```

```
df6_1 = df6_1[['order_id','vehicle_no']]
df6_1 = df6_1.head(10000)
```

```
df6_2 = pd.read_csv('New FleetJourney.csv')
print(df6_2.columns.tolist())
```

```
df6_2 = df6_2[['SENDERS_DATE','vehicleType', 'TRANSPORTATION_DISTANCE_IN_KM',]]
df6_2 = df6_2.head(10000)
```

```
# Concat 2 data frame 
df6 = pd.concat([df6_1, df6_2], axis = 1)
df6
```

```
# Change column "SENDERS_DATE" to datetime
df6['SENDERS_DATE'] = pd.to_datetime(df6['SENDERS_DATE'])

# Set period
start_date = pd.to_datetime('11/01/2021')
end_date = pd.to_datetime('30/11/2021')
# Filtered the data frame
df6_filtered = df6[(df6['SENDERS_DATE'] >= start_date) & (df6['SENDERS_DATE'] <= end_date)]
df6_filtered
```

```
# Tính tần suất xuất hiện của từng loại xe theo ngày tháng
vehicle_counts = df6_filtered.groupby(['SENDERS_DATE', 'vehicleType']).size().reset_index(name='Frequency')
print(vehicle_counts)
```

```
# Calculate the usage frequency of every date within the period
date_freq = df6_filtered['SENDERS_DATE'].value_counts().sort_index()
print(date_freq)
```

```
# Plot usage frequency to date
plt.bar(date_freq.index, date_freq.values)
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Usage frequency of vehicle in 11/2021')
plt.xticks(rotation=45)
plt.show()
```

```
#Calculate usage frequency of every vehicle type in the period 
type_frequency = df6_filtered['vehicleType'].value_counts()
# Top 5 vehicle 
top_5 = type_frequency.head(5)
top_5
```

```
vehicles = top_5.index.tolist()
#Plot line for every vehicle in top 5
for vehicle in vehicles:
    selected_vehicle = df6_filtered[df6_filtered['vehicleType'] == vehicle] #create new dataframe and select rows in 'df6_filtered' where 'vehicleType' has a value equals to 'vehicle'
    date_freq = selected_vehicle['SENDERS_DATE'].value_counts().sort_index() #calculates the frequency of values in the 'SENDERS_DATE' column within 'selected_vehicle'
    plt.plot(date_freq.index, date_freq.values, label=vehicle) #uses the index of 'date_freq' as the x-axis and the values of 'date_freq' as the y-axis. 

# Create plot of Top 5 Vehicle Usage frequency
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title('Top 5 Vehicle Usage Frequency in 11/2021')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

## Chart 7 : Comparison chart of the frequency of route usage by vehicles within the fleet during a period.

```
df7 = pd.read_csv('Route file.csv')
#Take necessary columns
df7 = df7[['order_id','SENDERS_DATE','RECEIVERS_DATE','Route_ID','vehicleType','PLACE_OF_SENDERS', 'PLACE_OF_LOADING', 'PLACE_OF_DISCHARGE']]
df7
```

```
# Create the Route detail (from senders' place to receivers' place)
df7['Route Detail'] = df7[['PLACE_OF_SENDERS','PLACE_OF_DISCHARGE']].apply(lambda x: '-'.join(x), axis=1)
df7
```

```
# Change column "SENDERS_DATE" to datetime
df7['SENDERS_DATE'] = pd.to_datetime(df7['SENDERS_DATE'])

# Set period
start_date = pd.to_datetime('2021/10/01')
end_date = pd.to_datetime('2021/12/31')
# Filtered the data frame
df7_filtered = df7[(df7['SENDERS_DATE'] >= start_date) & (df7['SENDERS_DATE'] <= end_date)]
df7_filtered
```

```
# Count the frequency of using the route
route_using = df7_filtered['Route Detail'].value_counts(sort=False)
route_using
```

```
# Create the bar plot
plt.figure(figsize= (12,8))
fig = route_using.plot(kind='bar', color= 'green')
plt.title("Route Usage Frequency in last quarter of 20221")
plt.xlabel("Route")
plt.ylabel("Frequency")

# Display the frequency and the route with highest frequency
for i, freq in enumerate(route_using):
    fig.text(i, freq, str(freq), ha='center', va='bottom')

# Display the plot
plt.xticks(rotation=90)
plt.show()

plt.savefig('Route Usage Frequency.png')
```
